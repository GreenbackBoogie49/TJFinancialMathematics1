```python
nifty_df = niftydata
mankind_df = ticker1data
cipla_df = ticker2data
sunph_df = ticker3data
drreddy_df = ticker4data
```

```python
nifty_df, mankind_df = nifty_df.align(mankind_df, join="inner", axis=0)
```

```python
nifty_df
```

```python
drreddy_df, mankind_df = drreddy_df.align(mankind_df, join="inner", axis=0)
```

```python
nifty_df = nifty_df[["Close", "Open"]]
```

```python
nifty_df
```

```python
mankind_df = mankind_df[["Close", "Open"]]
```

```python
nifty_df.columns = nifty_df.columns.droplevel(1)
```

```python
nifty_df.loc[:, "DailyReturns"] = (
    (nifty_df["Close"] - nifty_df["Open"]) / nifty_df["Open"]
)
```

```python
nifty_df
```

```python
mankind_df.columns = mankind_df.columns.droplevel(1)
```

```python
mankind_df.loc[:, "DailyReturns"] = (
    (mankind_df["Close"] - mankind_df["Open"]) / mankind_df["Open"]
)
```

```python
mankind_df
```

```python
drreddy_df.columns = drreddy_df.columns.droplevel(1)
drreddy_df.loc[:, "DailyReturns"] = (
    (drreddy_df["Close"] - drreddy_df["Open"]) / drreddy_df["Open"]
)
cipla_df.columns = cipla_df.columns.droplevel(1)
cipla_df.loc[:, "DailyReturns"] = (
    (cipla_df["Close"] - cipla_df["Open"]) / cipla_df["Open"]
)
```

```python
sunph_df.columns = sunph_df.columns.droplevel(1)
sunph_df.loc[:, "DailyReturns"] = (
    (sunph_df["Close"] - sunph_df["Open"]) / sunph_df["Open"]
)
```

```python
final_df = pd.concat([
    nifty_df["DailyReturns"].rename("NIFTY"),
    drreddy_df["DailyReturns"].rename("DRREDDY"),
    cipla_df["DailyReturns"].rename("CIPLA"),
    sunph_df["DailyReturns"].rename("SUNPHARMA"),
    mankind_df["DailyReturns"].rename("MANKIND")   
], axis = 1)
```

```python
final_df.head(5)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import root_mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
model = LinearRegression()
```

```python
#inputs X
X = final_df["NIFTY"].values.reshape((492,1))
```

```python
y_one =  final_df["DRREDDY"].values
```

```python
model.fit(X,y_one)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
# Scatter plot of actual data
plt.scatter(X, y_one)

# Create smooth x values for regression line
x_line = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)

# Predict y using the model
y_line = model.predict(x_line)

# Plot regression line
plt.plot(x_line, y_line)

plt.xlabel("Market Returns (NIFTY)")
plt.ylabel("Stock Returns (DR_REDDY)")
plt.title("CAPM Regression: Stock vs Market")

plt.show()
```

```python
X_train, X_test, y_one_train, y_one_test = train_test_split(
    X, y_one,
    test_size=0.2,
    random_state=42
)
```

```python
model.fit(X_train, y_one_train)
```

```python
new_predictions = model.predict(X_test)
```

```python
root_mean_squared_error(y_one_test, new_predictions)
```

```python
model.coef_
```
